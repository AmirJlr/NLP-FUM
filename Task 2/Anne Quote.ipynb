```python
# import packages:
from icrawler.builtin import GoogleImageCrawler

import warnings 
warnings.filterwarnings('ignore')

from pathlib import Path

from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw 

import re

import pytesseract
from pytesseract import image_to_string
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

from deep_translator import GoogleTranslator

import arabic_reshaper
from bidi.algorithm import get_display

from pynter.pynter import generate_captioned

from thefuzz import fuzz
from thefuzz import process
```

```python
# initialize :

max_num = 10
# assign directory
anne_dir = 'anne'    
wallpaper_dir = 'wallpapers'

# fonts!
font_path ='fonts/Roboto-Regular.ttf'

fontFile = "fonts/Sahel-Bold.ttf"

```

```python
def get_google_images(keyword, root_dir, max_num): 
    google_crawler = GoogleImageCrawler(storage={'root_dir': root_dir})
    google_crawler.crawl(keyword=keyword, max_num=max_num)
       
    files = Path(root_dir).glob('*')
    
    while len(list(files)) < max_num:
        get_google_images(keyword, root_dir, max_num)
    return files
           
```

```python
files = get_google_images('sky background', 'wallpapers', max_num)
```

```python
files = get_google_images('Top Anne shirley Quotes', 'anne', max_num)
```

```python
# iterate over files
anne_files = Path(anne_dir).glob('*')
wallpaper_files = Path(wallpaper_dir).glob('*')
counts = range(max_num)



for c, anne, wall in zip(counts, anne_files, wallpaper_files):
    text1 = re.sub('\n', ' ', image_to_string(Image.open(anne)))
    print(text1)
    print('----------')
    
    text2 = re.sub("[^a-zA-Z]", " ", text1)
    text2= re.sub('[^\S\r\n]{2,}', ' ', text2)
    print(text2)
    print('----------')
    
    translated = re.sub('[a-zA-Z]','', GoogleTranslator(source='auto', target='fa').translate(text2))
    print(translated)
    print('*********************************')

    # this was a 400x400 jpg file
    imageFile = wall
    image = Image.open(imageFile)

    # first you must prepare your text (you dont need this step for english text)
    text = translated
    reshaped_text = arabic_reshaper.reshape(text)    # correct its shape
    bidi_text = get_display(reshaped_text)           # correct its direction
    
    width  = image.width
    height = image.height
    img_center = (width/2,height/2)

    font = ImageFont.truetype(fontFile, 16)
    # start drawing on image
    draw = ImageDraw.Draw(image)
    draw.text((0, 0), bidi_text, (255,0,0), font=font, align='right')
    
    # save it
    image.save(f'imgs/{c}.jpg')
    
```

```python
def create_table(files1, files2, files3):
    header = """

<h1 style="text-align: center;">Anne Shirley Quotes</h1>

<table style="margin-left: auto; margin-right: auto;">
<thead>
<tr>
    <th>Original Image</th>
    <th>Wallpaper Image</th>
    <th>Final Image</th>
</tr>
</thead>


<tbody>
    """
    
    rows=""
    for f1, f2, f3 in zip(files1, files2, files3):

        rows += f"""
<br>  
<tr>
    <td><img src="{f1}" alt="" width="300px" height="300px"></td>
    <td><img src="{f2}" alt="" width="300px" height="300px"></td>
    <td><img src="{f3}" alt="" width="300px" height="300px"></td>
</tr>
<br>
    """
        
        
    footer = """
</tbody>
</table>
        """

    return header, rows, footer
```

```python
# iterate over files
anne_files = Path('anne').glob('*')
wallpaper_files = Path('wallpapers').glob('*')
imgs_files = Path('imgs').glob('*')

h, r, f =create_table(anne_files, wallpaper_files, imgs_files)
```
